```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = range(-9,11)
y = [x**2 - x + 2 for x in x]


plt.plot(x, y, marker=".")
plt.annotate("Orgin", xy=(0,0), xytext=(0,20), arrowprops={"facecolor": "black"})
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
x = range(0,100)
y = [x**.5 for x in x]
plt.plot(x, y, marker="|")

plt.subplot(2, 2, 2)
x = range(-9,10)
y = [x**3 for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 3)
x = range(-10,11)
y = [math.tan(x) for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 4)
x = range(-10,11)
y = [2**x for x in x]
plt.plot(x, y, marker=".")
```
