## Plotting $y=\tan(x)$ in Matplotlib

This approach:

- Plots many points between $-2\pi$ and $2\pi$ to increase the resolution of the data
- Handles asymtotes and removes the connecting line between $lim_{x\to-\infty}$ and $lim_{x\to\infty}$
- Plots the x axis ticks with radian measurements
- Makes me yearn for a TI-86 calculator
- Was excellent practice with matplotlib

```python
# import numpy to create a range with more specificity than incrementing by 1
import numpy as np

# 
x = np.linspace(-2 * math.pi, 2 * math.pi, 1000)

# np.tan(x) is similar to [math.tan(n) for n in x]
y = np.tan(x)

# give me everything other than the last element
# looking at difference between two numbers
# where i have a difference between a posistive and negative number,
# do not plot it
y[:-1][np.diff(y) < 0] = np.nan # comment this line out if we want to see the tangent of x

# show grid
plt.grid()

plt.xlabel("x")
plt.ylabel("$tan(x)$")

plt.ylim(-10, 10)
plt.xlim(-2 * math.pi, 2 * math.pi)

# x_labels in radians
# For a more programmatic approach to radians,
# see https://matplotlib.org/3.1.1/gallery/units/radian_demo.html
radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']

plt.xticks(radians, radian_labels)

plt.title("$y = tan(x)$")
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
x = range(0,100)
y = [x**.5 for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 2)
x = range(-9,10)
y = [x**3 for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 3)
x = range(-10,11)
y = [math.tan(x) for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 4)
x = range(-10,11)
y = [2**x for x in x]
plt.plot(x, y, marker=".")
plt.show
```

```python
x1 = range(-10,11)
y1 = [x**.5 for x in x]

x2 = range(-10,11)
y2 = [x**3 for x in x]

x3 = range(-10,11)
y3 = [math.tan(x) for x in x]

x4 = range(-10,11)
y4 = [2**x for x in x]

plt.figure(figsize=(16, 9))

plt.plot(x1, y1, marker=".", label="square root")
plt.plot(x2, y2, marker=".", label="cube")
plt.plot(x3, y3, marker=".", label="tan")
plt.plot(x4, y4, marker=".", label="exponent")
plt.legend()
```
