```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
x = range(-9,11)
y = [x**2 - x + 2 for x in x]


plt.plot(x, y, marker=".")
plt.annotate("Orgin", xy=(0,0), xytext=(0,20), arrowprops={"facecolor": "black"})
```

```python
x = range(0,100)
y = [x**.5 for x in x]

plt.plot(x, y)
```

```python
x = range(-9,10)
y = [x**3 for x in x]

plt.plot(x, y)
```

```python
x = range(-199,200)
y = [math.tan(x) for x in x]

plt.plot(x, y)
```

```python
x = range(-199,200)
y = [2**x for x in x]

plt.plot(x, y)
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
x = range(0,100)
y = [x**.5 for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 2)
x = range(-9,10)
y = [x**3 for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 3)
x = range(-10,11)
y = [math.tan(x) for x in x]
plt.plot(x, y, marker=".")

plt.subplot(2, 2, 4)
x = range(-10,11)
y = [2**x for x in x]
plt.plot(x, y, marker=".")
```

```python
x1 = range(-10,11)
y1 = [x**.5 for x in x]

x2 = range(-10,11)
y2 = [x**3 for x in x]

x3 = range(-10,11)
y3 = [math.tan(x) for x in x]

x4 = range(-10,11)
y4 = [2**x for x in x]

plt.figure(figsize=(16, 9))

plt.plot(x1, y1, marker=".", label="square root")
plt.plot(x2, y2, marker=".", label="cube")
plt.plot(x3, y3, marker=".", label="tan")
plt.plot(x4, y4, marker=".", label="exponent")
plt.legend()
```
