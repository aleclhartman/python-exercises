```python
# 4. Which features would be best used to predict species?
plt.figure(figsize=(16, 9))
plt.suptitle("Which features would be best used to predict species? (Represented in boxplots)")

plt.subplot(221)
sns.boxplot(data=df, x="sepal_width", y="species")
plt.xlabel("Sepal Width")
plt.ylabel("Species")
plt.yticks(rotation=90)
plt.title("Sepal Width by Species")

plt.subplot(222)
sns.boxplot(data=df, x="sepal_length", y="species")
plt.xlabel("Sepal Length")
plt.ylabel("Species")
plt.yticks(rotation=90)
plt.title("Sepal Length by Species")

plt.subplot(223)
sns.boxplot(data=df, x="petal_width", y="species")
plt.xlabel("Petal Width")
plt.ylabel("Species")
plt.yticks(rotation=90)
plt.title("Petal Width by Species")

plt.subplot(224)
sns.boxplot(data=df, x="petal_length", y="species")
plt.xlabel("Petal Length")
plt.ylabel("Species")
plt.yticks(rotation=90)
plt.title("Petal Length by Species")

plt.subplots_adjust(hspace=.4)
```

```python
# feature engineering aside
df["sepal_area"] = df["sepal_length"] * df["sepal_width"]
df["petal_area"] = df["petal_length"] * df["petal_width"]
df["sepal_to_petal"] = df["sepal_area"] / df["petal_area"]
feature_engineering = df[["sepal_area", "petal_area", "sepal_to_petal", "species"]]
```

```python
sns.pairplot(data=feature_engineering, hue="species")
plt.show()
```

```python
# 4. Using the chipotle dataset from the previous exercise,
def get_db_url(database):
    url = "mysql+pymysql://{}:{}@{}/{}"
    return url.format(user, password, host, database)

chipotle_url = get_db_url("chipotle")
```

```python
chipotle = pd.read_sql("""SELECT *
FROM orders""", chipotle_url)
chipotle
```

```python
chipotle["item_price"] = chipotle["item_price"].str.replace("$", "").str.replace(",", "").astype(float)
```

```python
chipotle.head()
```

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
four_most_popular_items_quantity = chipotle.groupby("item_name").quantity.sum().sort_values(ascending=False).head(4)
four_most_popular_items_quantity
```

```python
four_most_popular_items_revenue = chipotle.groupby("item_name").item_price.sum().sort_values(ascending=False).head(4)
four_most_popular_items_revenue
```

```python
top_four_items = pd.DataFrame(four_most_popular_items_revenue).reset_index()
top_four_items
```

```python
sleepstudy = data("sleepstudy")
sleepstudy
```

```python
sleepstudy.groupby("Subject").Days.count()
```

```python
colorblind = sns.palplot(sns.color_palette("colorblind"))
```

```python
# Use seaborn to create a line chart of all the individual subject's reaction times and
# a more prominant line showing the average change in reaction time.
plt.figure(figsize=(16,9))
sns.lineplot(x="Days", y="Reaction", data=sleepstudy, color=sns.color_palette("colorblind")[0], ci="sd")
sns.lineplot(x="Days", y="Reaction", data=sleepstudy, palette="Paired", hue="Subject", legend="full", alpha=.5)
plt.title("Change in Reaction Time as Days of Sleep Deprivation Increase")
plt.ylabel("Reaction Time (in Milliseconds)")
plt.legend(loc="best", bbox_to_anchor=(1.1, 0.8), shadow=True, ncol=1)
plt.show()
```
