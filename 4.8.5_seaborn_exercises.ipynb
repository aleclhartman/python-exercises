```python
# 4. Using the chipotle dataset from the previous exercise,
def get_db_url(database):
    url = "mysql+pymysql://{}:{}@{}/{}"
    return url.format(user, password, host, database)

chipotle_url = get_db_url("chipotle")
```

```python
chipotle = pd.read_sql("""SELECT *
FROM orders""", chipotle_url)
chipotle
```

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
four_most_popular_items_quantity = chipotle.groupby("item_name").quantity.sum().nlargest(4)
four_most_popular_items_quantity
```

chipotle.groupby("item_name")

```python
four_most_popular_item_names = [item for item in chipotle["item_name"] if chipotle.groupby("item_name").quantity.sum()[1] == four_most_popular_items_quantity[1]]
four_most_popular_item_names
```

```python
chipotle["item_price"] = chipotle["item_price"].str.replace("$", "").str.replace(",", "").astype(float)
chipotle
```

###### Programmitcally determine xlabels

```python
plt.figure(figsize=(16,9))
sns.barplot(data=chipotle, x=["Chicken Bowl", "Chicken Burrito", "Chips and Guacamole", "Steak Burrito"] , y=four_most_popular_items_quantity)
plt.xlabel("Items")
plt.ylabel("Quantity")
plt.title("Four Most Popular Items at Chipotle")
plt.text(
    0,
    769,
    f'Revenue ${chipotle[chipotle.item_name == "Chicken Bowl"].item_price.sum().round(2)}'
)
plt.text(
    1,
    600,
    f'Revenue ${chipotle[chipotle.item_name == "Chicken Burrito"].item_price.sum().round(2)}'
)
plt.text(
    2,
    515,
    f'Revenue ${chipotle[chipotle.item_name == "Chips and Guacamole"].item_price.sum().round(2)}'
)
plt.text(
    3,
    395,
    f'Revenue ${chipotle[chipotle.item_name == "Steak Burrito"].item_price.sum().round(2)}'
)
plt.show()
```

```python
# Use seaborn to create a line chart of all the individual subject's reaction times and
# a more prominant line showing the average change in reaction time.
plt.figure(figsize=(16,9))
sns.lineplot(x="Days", y="Reaction", data=sleepstudy, marker="|", color=sns.color_palette("colorblind")[2])
sns.lineplot(x="Days", y="Reaction", data=sleepstudy, palette="Paired", hue="Subject", marker=".", legend="full")
plt.title("Change in Reaction Time as Days of Sleep Deprivation Increase")
plt.ylabel("Reaction Time (in Milliseconds)")
plt.show()
```
