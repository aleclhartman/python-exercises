```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pydataset import data
```

```python
# Use the iris database to answer the following quesitons:
df = data("iris")
df.head()
```

```python
df.rename(columns={"Sepal.Length": "sepal_length"}, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={
    "Sepal.Width": "sepal_width",
    "Petal.Length": "petal_length",
    "Petal.Width": "petal_width",
    "Species": "species"
}, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df["petal_length"].value_counts().head()
```

```python
# 1. What does the distribution of petal lengths look like?
sns.distplot(df["petal_length"])
plt.xlabel("Petal Length")
plt.ylabel("Probability Density")
plt.title("Distribution of Petal Lengths")
plt.show()
```

```python
sns.distplot(df["petal_length"], kde=False)
plt.xlabel("Petal Length")
plt.ylabel("Count")
plt.title("Histogram of Petal Lengths")
plt.show()
```

```python
sns.distplot(df["petal_length"], hist=False)
plt.xlabel("Petal Length")
plt.ylabel("Probability Density")
plt.title("Kernel Density Estimation of Petal Lengths")
plt.show()
```

```python
# 2. Is there a correlation between petal length and petal width?
df.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.corr(), annot=True, cmap="Greens")
plt.show()
```
