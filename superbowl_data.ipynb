Does the location of the game impact the outcome?

In essence, if the city where the game is held is closer to one team's city over the other, is the team from the city nearest to the event more likely to win?

```python
superbowl = pd.read_csv("superbowl.csv")
superbowl.sample(10)
```

```python
superbowl["total_points"] = superbowl["Winner Pts"] + superbowl["Loser Pts"]
superbowl.sample(5)
```

```python
sns.palplot(sns.color_palette("colorblind"))
```

```python
sns.relplot(x="useful_date", y="total_points", data=superbowl)
plt.xlabel("Years")
plt.ylabel("Points")
plt.title("Total Points Over Time")
plt.show()
```

```python
sns.boxplot(y="total_points", data=superbowl)
```

```python
sns.pairplot(superbowl)
```
