# matplotlib

Primarily work in jupyter notebooks for matplotlb

try %matplotlib notebook

Plot list of numbers
x and y need to have same number of data points


```python
%matplotlib inline


import matplotlib.pyplot as plt
```

```python
plt
```

```python
y = [1, 2, 3, 4, 5]

plt.plot(y)
```

```python
# y = x^2
x = range(-4, 5)
y = [x**2 for x in x]

plt.plot(x, y, marker='x')
```

```python
x = ["data science", "web dev"]
y = [16, 7]

plt.bar(x, y)
plt.ylabel("Avg Student Coolness")
plt.title("Avg Coolness by Codeup Program")
```

```python
x = ["data science", "web dev"]
y = [16, 7]

plt.barh(x, y, edgecolor='blue', facecolor='yellow')
plt.xlabel("Avg Student Coolness")
plt.title("Avg Coolness by Codeup Program")
```
