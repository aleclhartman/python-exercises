### 5 components of a program

```python
x = np.arange(-1000, 1000)
```

```python
mask = x > 950
```

is each element greater than 950?

```python
mask
```

```python
x[mask]
```

```python
# how many numbers match the mask
mask.sum()
```

```python
# sum all the numbers that match the mask (x > 950)
x[mask].sum()
```

```python
x[mask].mean()
```

```python
higher_than_950 = x[mask]
```

```python
higher_than_950.max()
```

```python
higher_than_950.min()
```

```python
higher_than_950.argmax() # argmax returns the index of the highest number
```

```python
higher_than_950[higher_than_950.argmax()]
```

```python
x = np.arange(-1000, 1000)
x[(x > 5) & (x % 2 == 0)] # single & for plural ands
```

```python
x = np.arange(-1000, 1000)
x[(x > 0) | (x % 2 == 0)] # single & for plural ands
```

```python
x = np.arange(-1000, 1000)
x = x[x > 0] # x where x is greater than 0
x = x[x % 2 == 0] # read this as select x where x > 0
```
