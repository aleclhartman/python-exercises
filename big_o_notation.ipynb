```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,11)
y1 = [x*0 for x in x]

y2 = [math.log(x) for x in x]

y3 = [x for x in x]

y4 = [x * math.log(x) for x in x]

y5 = [x**2 for x in x]

y6 = [2**x for x in x]

y7 = [math.factorial(x) for x in x]

y8 = [x**x for x in x]

plt.figure(figsize=(16, 9))
plt.ylim(-5,100)

plt.xlabel("Elements")
plt.ylabel("Operations")
plt.title("Big O Notation")

plt.plot(x, y1, marker=".", label="0(1)")
plt.plot(x, y2, marker=".", label="0log(n)")
plt.plot(x, y3, marker=".", label="0(n)")
plt.plot(x, y4, marker=".", label="0(n log n)")
plt.plot(x, y5, marker=".", label="0(n^2)")
plt.plot(x, y6, marker=".", label="0(2^n)")
plt.plot(x, y7, marker=".", label="0(n!)")
plt.plot(x, y8, marker=".", label="0(n^n)")
plt.legend()
```
